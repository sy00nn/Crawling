**서울공공데이터**

```python
import requests
from bs4 import BeautifulSoup

# 시작 페이지와 끝 페이지 설정
start_page = 1
end_page =int(input('끝 페이지 입력하시오: '))

# 페이지 순회하며 데이터 수집
for current_page in range(start_page, end_page + 1):
    url = f"https://data.seoul.go.kr/dataList/datasetList.do?datasetKind=1&searchFlag=M&pageIndex={current_page}"
    page = requests.get(url)

    # 응답 상태 코드 확인
    if page.status_code == 200:
        try:
            # BeautifulSoup으로 HTML 파싱
            soup = BeautifulSoup(page.text, 'html.parser')

            # 필요한 데이터가 들어있는 dl 태그들 찾기 (태그 구조에 맞게 변경)
            datasets = soup.find_all('dl', {'class': 'type-b'})

            # 각 데이터셋 정보 추출 및 화면 출력
            for dataset in datasets:
                # 타이틀 추출
                title_tag = dataset.find('strong')
                title = title_tag.text.strip() if title_tag else "N/A"

                # 정보 추출
                description_tag = dataset.find('dd', {'class': 'list-statistics-info1'})
                description = description_tag.text.strip() if description_tag else "N/A"

                # 수정일자와 제공기관은 'list-statistics-info2' 클래스 아래에서 찾음
                info_section = dataset.find('dd', {'class': 'list-statistics-info2'})

                # 수정일자 추출
                if info_section:
                    date_b_tag = info_section.find('b', text=lambda t: '수정일자' in t)
                    if date_b_tag and date_b_tag.next_sibling:
                        date = date_b_tag.next_sibling.strip()
                    else:
                        date = "N/A"
                else:
                    date = "N/A"

                # 제공기관 추출
                if info_section:
                    region_b_tag = info_section.find('b', text=lambda t: '제공기관' in t)
                    if region_b_tag and region_b_tag.next_sibling:
                        region = region_b_tag.next_sibling.strip()
                    else:
                        region = "N/A"
                else:
                    region = "N/A"

                # 데이터 출력
                print(f"타이틀: {title}")
                print(f"정보: {description}")
                print(f"날짜: {date}")
                print(f"지역: {region}")
                print("-" * 40)

        except Exception as e:
            print(f"페이지 {current_page}에서 데이터 처리 오류 발생: {e}")
    else:
        print(f"페이지 {current_page} 요청 실패. 상태 코드: {page.status_code}")
```

**csv 파일 저장 코드**

```python
import requests
from bs4 import BeautifulSoup
import csv

# 시작 페이지와 끝 페이지 설정
start_page = 1
end_page = int(input('끝 페이지 입력하시오: '))

# CSV 파일 생성
with open('/content/drive/MyDrive/crawl_seoul_dateset.csv', 'w', newline='', encoding='UTF-8') as f:
    wr = csv.writer(f)

    # CSV 헤더 작성
    wr.writerow(['타이틀', '정보', '날짜', '지역'])

    # 페이지 순회하며 데이터 수집
    for current_page in range(start_page, end_page + 1):
        url = f"https://data.seoul.go.kr/dataList/datasetList.do?datasetKind=1&searchFlag=M&pageIndex={current_page}"
        page = requests.get(url)

        # 응답 상태 코드 확인
        if page.status_code == 200:
            try:
                # BeautifulSoup으로 HTML 파싱
                soup = BeautifulSoup(page.text, 'html.parser')

                # 필요한 데이터가 들어있는 dl 태그들 찾기 (태그 구조에 맞게 변경)
                datasets = soup.find_all('dl', {'class': 'type-b'})

                # 각 데이터셋 정보 추출 및 CSV 저장
                for dataset in datasets:
                    # 타이틀 추출
                    title_tag = dataset.find('strong')
                    title = title_tag.text.strip() if title_tag else "N/A"

                    # 정보 추출
                    description_tag = dataset.find('dd', {'class': 'list-statistics-info1'})
                    description = description_tag.text.strip() if description_tag else "N/A"

                    # 수정일자와 제공기관은 'list-statistics-info2' 클래스 아래에서 찾음
                    info_section = dataset.find('dd', {'class': 'list-statistics-info2'})

                    # 수정일자 추출
                    if info_section:
                        date_b_tag = info_section.find('b', text=lambda t: '수정일자' in t)
                        if date_b_tag and date_b_tag.next_sibling:
                            date = date_b_tag.next_sibling.strip()
                        else:
                            date = "N/A"
                    else:
                        date = "N/A"

                    # 제공기관 추출
                    if info_section:
                        region_b_tag = info_section.find('b', text=lambda t: '제공기관' in t)
                        if region_b_tag and region_b_tag.next_sibling:
                            region = region_b_tag.next_sibling.strip()
                        else:
                            region = "N/A"
                    else:
                        region = "N/A"

                    # CSV 파일에 데이터 작성
                    wr.writerow([title, description, date, region])

            except Exception as e:
                print(f"페이지 {current_page}에서 데이터 처리 오류 발생: {e}")
        else:
            print(f"페이지 {current_page} 요청 실패. 상태 코드: {page.status_code}")

print("데이터 수집 및 CSV 저장 완료")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 크롬 옵션 설정
chrome_options = Options()
chrome_options.add_argument('--headless')  # UI 없이 실행
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# 크롬 드라이버 실행
driver = webdriver.Chrome(options=chrome_options)
driver.get('https://finance.naver.com/main/main.nhn')

# 페이지 로드를 기다리기 위한 시간 (네트워크 상태에 따라 변경 가능)
time.sleep(3)

# 탭들의 XPath 리스트 (거래상위, 상승, 하락, 시가총액 상위 탭)
tabs = [
    '//*[@class="tab1"]/a',  # 거래상위
    '//*[@class="tab2"]/a',  # 상승
    '//*[@class="tab3"]/a',  # 하락
    '//*[@class="tab4"]/a'   # 시가총액 상위
]

# 각 탭을 클릭하여 활성화된 데이터 추출
for index, tab_xpath in enumerate(tabs):
    # 각 탭 클릭
    driver.find_element(By.XPATH, tab_xpath).click()

    # 탭 클릭 후 페이지 갱신을 위한 잠깐 대기
    time.sleep(2)

    # 활성화된 div에서 데이터를 찾기
    tbody_id = f"_topItems{index+1}"  # _topItems1, _topItems2, _topItems3, _topItems4에 대응
    try:
        # 활성화된 tbody 요소 찾기
        tbody = driver.find_element(By.ID, tbody_id)
        rows = tbody.find_elements(By.TAG_NAME, 'tr')

        # 각 행(tr)에서 데이터 추출
        for row in rows:
            cols = row.find_elements(By.TAG_NAME, 'td')
            stock_name = row.find_element(By.TAG_NAME, 'th').text  # 종목명은 <th> 안에 있음
            current_price = cols[0].text  # 현재가
            change = cols[1].text  # 전일 대비
            fluctuation = cols[2].text  # 등락률

            # 출력 (혹은 원하는 대로 데이터 저장)
            print(f"종목명: {stock_name}, 현재가: {current_price}, 전일대비: {change}, 등락률: {fluctuation}")

    except Exception as e:
        print(f"Error while scraping data for tab {index+1}: {e}")

# 크롬 드라이버 종료
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 크롬 옵션 설정
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')  # UI 없이 실행
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# 크롬 드라이버 실행
driver = webdriver.Chrome(options=chrome_options)
wait = WebDriverWait(driver, 10)

# URL로 이동
driver.get("https://www.open.go.kr/com/tema/temaSrhList.do")
driver.set_window_size(1920, 1080)

# 각 페이지에서 <ul> 요소를 크롤링하는 함수
def crawl_page(page_num):
    ul_elements = driver.find_elements(By.XPATH, "//div[@class='info_list']//ul")
    print(f"페이지 {page_num}")
    for ul in ul_elements:
        print(ul.text)  # 원하는 방식으로 출력하거나 저장

# 카테고리 순차적으로 탐색
category_xpath = "//td/a[@class='tyinfo']"

for category_index in range(1, 7):  # 각 카테고리의 XPath 인덱스 변경 필요
    # 카테고리 리스트 재조회
    category_elements = driver.find_elements(By.XPATH, category_xpath)

    # 카테고리 선택
    category = category_elements[category_index - 1]
    category_name = category.text
    print(f"\n카테고리: {category_name}")

    # 카테고리 클릭
    category.click()
    time.sleep(2)  # 페이지 로딩 대기

    # 페이지 1~10 크롤링
    for page_num in range(1, 11):
        crawl_page(page_num)

        # 다음 페이지로 이동
        try:
            next_page = wait.until(EC.element_to_be_clickable((By.LINK_TEXT, str(page_num + 1))))
            next_page.click()
            time.sleep(2)  # 페이지가 로드될 때까지 대기
        except Exception as e:
            print(f"페이지 {page_num + 1}로 이동하는 중 오류 발생: {e}")
            break

    # 처음 페이지로 다시 돌아가기
    driver.back()
    time.sleep(2)

# 크롤링이 끝나면 드라이버 종료
driver.quit()
```
